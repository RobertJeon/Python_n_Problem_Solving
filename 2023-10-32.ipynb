```python
# 고릴라 클래스를 구현해서 고릴라 객체를 만드시오!

"""
소리 지를 때마다 바나나 하나씩 소진
바나나가 다 소진 되었는데 또 소리를 지르려고 하면
"배가고파 소리를 지를 수 없습니다." 메세지가 출력되게하시오~


명령예시
gorila_1 = Gorilla()
gorila_1.eat(5)

출력예시
바나나를 5개를 먹었습니다.

명령예시
gorila_1.cry(3)

출력예시
우와 ~~~~우와 ~~~~우와 ~~~~
"""

"""
상속 및 메소드 오버라이딩 활용

고릴라 클래스 를 상속하여 킹콩 클래스 를 만들어봅니다.

킹콩은 바나나를 먹을때 2배의 효과를 냅니다.
"""

# class Person:
#     def __init__(self, name):
#         self.name = name

# class Student(Person):      # Person 클래스 상속받음(name 변수를 파라미터로 재사용)
#     def study(self):
#         print(self.name + " studies hard")

# class Employee(Person):     # Person 클래스 상속받음(name 변수를 파라미터로 재사용)
#     def work(self):
#         print(self.name + " works hard")

# s = Student("Dave")
# e = Employee("David")

# s.study()
# e.work()

# gorila_1 = Gorilla()gorila_1.eat(5)
# gorila_1.cry(3)

# def func2(n):
#     for i in range(n):
#         print("{} 번째 출력".format(i))

class Gorilla:
    def __init__(self):
        self.banana = 10  # 초기 바나나 개수는 10으로 설정합니다.

    def eat(self, count):
        if self.banana >= count:
            for i in range(count):
                print("우와 ~~~")
                self.banana -= count
            
            print(f"바나나를 {count}개를 먹었습니다.")
        else:
            print("배가고파 소리를 지를 수 없습니다.")


# 고릴라 객체 생성
gorilla_1 = Gorilla()

# 테스트
gorilla_1.eat(5)
gorilla_1.cry(3)



```
